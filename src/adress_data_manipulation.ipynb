```python
def transform_number_under_10(x):
    if int(x) < 10:
        x = '0' + str(int(x))
        return x
    else:
        x = str(int(x))
        return x
```

```python
total_df.drop(columns=total_df.columns[2:7], inplace = True)
total_df
```

```python
columns = total_df.columns

# for col in columns:
#     print(total_df[col].value_counts())

mybib.nullTable(total_df)

column_to_transform = ['7', '8', '9', '10']
for column in column_to_transform:
    total_df[column] = total_df[column].apply(lambda x: transform_number_under_10(x))
    
total_df

total_df.drop_duplicates(inplace=True)
total_df
# filter = total_df[total_df['8'].isnull()]
# display(filter)

```
