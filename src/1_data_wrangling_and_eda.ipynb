```python

accidents['ac_category'] = accidents['ac_category'].replace({1:'deadly', 2:'seriously_injured', 3:'light_injured' })
accidents['ac_type'] = np.where(accidents['ac_type'].isin([1, 4, 5, 6, 7, 8, 9, 0]), 'other', 'turn_in')

print(accidents.ac_category.value_counts())
print(accidents.ac_type.value_counts())
print(accidents.year.value_counts())
print(accidents.lor.value_counts())


city_center = (52.51916459, 13.405665044)

# https://www.section.io/engineering-education/using-geopy-to-calculate-the-distance-between-two-points/
# !pip install geopy
from geopy.distance import geodesic as GD


def get_distance(row):
    latitude = row['latitude']
    longitude = row['longitude']
    point = (latitude, longitude)
    city_center = (52.51916459, 13.405665044)
    distance = GD(point, city_center).km
    return distance 

accidents['distance_to_CC'] = accidents.apply(get_distance, axis =1)
accidents
```

```python
sum = final_df.groupby(['lor']).agg({'key':pd.Series.nunique, 'distance_to_CC':np.mean}).reset_index()
sum.columns = ['lor', 'total_accident_per_lor', 'avg_distance_to_CC_per_lor']
sum['total_accident_per_lor'] = sum['total_accident_per_lor'].astype(int)
sum
```

```python
summary_table = pd.merge(summary_table, sum, on = 'lor', how = 'left')
summary_table['population_density_per_lor_in_sqkm'] = round(summary_table['total_population'] / summary_table['area_in_sqm'] * 1000000,0)

display(summary_table)
summary_table.columns
```

```python
mybib.nullTable(summary_table)
summary_table['total_accident_per_lor'].fillna(0, inplace=True)
```

```python
summary_table.to_csv('../data/output/Tableau/final_summary_table.csv', index=False)
summary_table.to_excel('../data/output/Tableau/final_summary_table.xlsx', index=False)
summary_table
```
