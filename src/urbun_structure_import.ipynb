```python
urban_data = pd.read_excel('../data/input/urban_structure/urban_structure.xlsx')
```

```python
# https://www.geeksforgeeks.org/apply-function-to-every-row-in-a-pandas-dataframe/
  
urban_data['Flächennutzung (Code)'] = urban_data.apply(lambda row: row['Grün- und Freiflächennutzung (Code)'] if row['Nutzung der bebauten Flächen (Code)'] == '-' else row['Nutzung der bebauten Flächen (Code)'], axis=1)

urban_data['Flächennutzung (Code)'].value_counts()

urban_data['Flächennutzung'] = urban_data.apply(lambda row: row['Grün- und Freiflächennutzung'] if row['Nutzung der bebauten Flächen'] == '-' else row['Nutzung der bebauten Flächen'], axis=1)

urban_data['Flächennutzung'].value_counts()
```

```python
# select 2 columns - drop duplicates - create index table - rename columns

land_use_index = urban_data[['Flächennutzung (Code)','Flächennutzung']]
land_use_index = land_use_index.drop_duplicates().sort_values(by=['Flächennutzung (Code)']).reset_index(drop = True)
land_use_index.columns = ['land_use_id', 'land_use']
display(land_use_index)

land_use_index.to_csv(f'../data/output/land_use_index.csv', index = True)
land_use_index.to_pickle(f'../data/output/land_use_index.pkl')
```

```python
# fill empty values in column
urban_data['Stadtstrukturtyp (Code)'] = urban_data.apply(lambda row: int(0) if row['Stadtstrukturtyp (Code)'] == '-' else row['Stadtstrukturtyp (Code)'], axis=1)

# select 2 columns - drop duplicates - create index table - rename columns

city_structure = urban_data[['Stadtstrukturtyp (Code)','Stadtstrukturtyp']]
city_structure = city_structure.drop_duplicates().sort_values(by = 'Stadtstrukturtyp (Code)').reset_index(drop = True)
city_structure.columns = ['structure_type_id', 'structure_type']

display(city_structure)

city_structure.to_csv(f'../data/output/city_structure_type_index.csv', index = True)
city_structure.to_pickle(f'../data/output/city_structure_type_index.pkl')
```

```python
# select 2 columns - drop duplicates - create index table - rename columns
district_index = urban_data[['Bezirk (Code)','Bezirk']]
district_index = district_index.drop_duplicates().sort_values(by = 'Bezirk (Code)').reset_index(drop = True)
district_index.columns = ['district_id', 'district']
display(district_index)

district_index.to_csv(f'../data/output/district_index.csv', index = True)
district_index.to_pickle(f'../data/output/district_index.pkl')
```
