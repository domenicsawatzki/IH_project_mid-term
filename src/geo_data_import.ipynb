```python

path = '../data/input/geo_data/lor_planungsraeume_2021.geojson'

# using utf-8 to handle ä,ü,ö,ß 
with open(path, 'r', encoding='utf-8') as f:
    gdf = gpd.read_file(f)
    
    
display(gdf)
```

```python
# https://stackoverflow.com/questions/20474549/extract-points-coordinates-from-a-polygon-in-shapely


gdf['coordinates'] = gdf['geometry'].apply(lambda geom: list(geometry.exterior.coords))
gdf['GROESSE_M2'] = gdf['GROESSE_M2'].apply(lambda x: round(x,2))

gdf.columns = ['lor', 'sub_district_name', 'district_id', 'last_updated', 'area_in_sqm', 'geometry', 'coordinates']

gdf 
```
